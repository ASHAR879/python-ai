**Aggregation**

in the context of data analysis, particularly when using pandas in Python, refers to the process of combining multiple pieces of data to produce a summary result. This process often involves applying a function or set of functions to a dataset to generate a single value or a few values that capture key aspects of the dataset.
Key Points about Aggregation:

**Purpose:**

Aggregation is used to summarize and condense data. It helps in deriving meaningful insights from large datasets by reducing the data into a more digestible form.

**Common Aggregation Functions:**

Some common aggregation functions include:

`Sum (sum):` Adds up all the values.

`Mean (mean):` Calculates the average of the values.

`Median (median):` Finds the middle value in the dataset.

`Min (min):` Identifies the smallest value.

`Max (max):` Identifies the largest value.

`Count (count):` Counts the number of values.

`Standard Deviation (std):` Measures the amount of variation or dispersion.

```python
df.agg(lambda x: np.mean(x) *5.6)
```

```python
df.transform(lambda x:x*101.2)
```

# Grouping

```python
df = pd.DataFrame(
    {
        "A" : ["foo","bar","foo","bar","foo","bar","foo","foo"],
        "B" : ["one","one","two","three","two","two","one","three"],
        "C" : np.random.randn(8),
        "D" : np.random.randn(8)
    }
)
df
```

```python
# it sum all values of C having label "bar" and "foo"
df.groupby("A")[["C","D"]].sum()
```

```python
df.groupby(["A","B"]).sum()
```
