# 16- splitting a string into multiple columns

```python
import pandas as pd

df = pd.DataFrame({'name':['Arsalan Akram','Asif Iqbal','Muwahid Khan'],'location':["Lahore, Pakistan", "Sargodha, Pakistan","Islamabad, Pakistan"]})
df
```

```python
df.name.str.split(' ').head()
```

```python

df.location.str.split(',').head()
```

```python
# expand first and last part of name into new column
df.name.str.split(' ',expand = True)
```

```python
df
```

```python
# Adding those split into new columns
df[["first_name","last_name"]]=df.name.str.split(' ',expand = True)
df
```

```python
# Expand location
df.location.str.split(' ',expand = True)
```

```python
# Adding those split into new columns
df[["city","country"]]=df.location.str.split(',',expand = True)
df
```

```python
# Refine Data manipulation
df = df[['first_name','last_name','city','country']]
df
```
