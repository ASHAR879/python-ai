# show that how many Mbs are in Kbs and how to calculate

2. remove M from all numbers
3. handle varies with device



```python
df['Size'].isnull().sum()
```

```python
import pandas as pd

# Convert 'K' to 'M' by dividing by 1024
def convert_size(size):
    if 'K' in size:
        # Convert from K to M
        return str(round(float(size.replace('K', '')) / 1024, 3)) + 'M'
    else:
        return size

# Apply the conversion function to the 'Size' column
df['Size'] = df['Size'].apply(convert_size)

```

```python
import pandas as pd
import numpy as np

# Convert 'Size' to float, handle non-numeric and numeric entries
def convert_size(size):
    # Check if the value is a string
    if isinstance(size, str):
        # Attempt to remove 'M' and convert to float
        try:
            return float(size.replace('M', ''))
        except ValueError:  # Handles any conversion errors for strings
            return np.nan  # Replace non-convertible strings with NaN
    else:
        # Directly return the value if it's already a number
        return size

# Apply the conversion function to the 'Size' column
df['Size'] = df['Size'].apply(convert_size)

df
```
